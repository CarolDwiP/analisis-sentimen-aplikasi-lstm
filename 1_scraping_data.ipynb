```python
import pandas as pd
from google_play_scraper import reviews, Sort
import time

# Konfigurasi target aplikasi
# Menggunakan beberapa aplikasi e-wallet dan e-commerce populer untuk memperbanyak variasi data
app_packages = [
    'com.instagram.android',
    'com.whatsapp',
    'com.shopee.id',
    'com.tokopedia.tkpd',
    'com.gojek.app',
    'com.telkom.tracencare'
]

data_mentah = []

# Fungsi scraping sederhana
def ambil_data_review(app_id, jumlah_target):
    hasil_reviews, _ = reviews(
        app_id,
        lang='id',
        country='id',
        sort=Sort.NEWEST,
        count=jumlah_target
    )
    return hasil_reviews

# Proses pengambilan data
# Target 1000 data per aplikasi untuk memastikan total dataset > 3000
target_per_app = 1000

print("Memulai proses scraping data...")

for app in app_packages:
    print(f"Mengambil data dari: {app}")
    try:
        review_data = ambil_data_review(app, target_per_app)
        for ulasan in review_data:
            data_mentah.append({
                'source': app,
                'username': ulasan['userName'],
                'rating': ulasan['score'],
                'review_text': ulasan['content'],
                'date': ulasan['at']
            })
        print(f"Berhasil mengambil {len(review_data)} ulasan.")
        time.sleep(2) # Delay untuk menghindari rate limit
    except Exception as e:
        print(f"Gagal mengambil data {app}: {e}")

# Konversi ke DataFrame
df_hasil = pd.DataFrame(data_mentah)

# Simpan ke CSV
nama_file = 'dataset_mentah_playstore.csv'
df_hasil.to_csv(nama_file, index=False)

print(f"Proses selesai. Total data terkumpul: {len(df_hasil)} baris.")
print(f"Data disimpan dalam file: {nama_file}")
```
